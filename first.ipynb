```python
marks = float(input("enter marks: "))

if marks >=80 and marks <=100:
    print("distinction")

elif marks >=60 and marks<=79.9:
    print("1st division")

elif marks >=50 and marks<=59.9:
    print("2nd division")

elif marks >=40 and marks<=49.9:
    print("3rd division")

elif marks<=39:
    print("fail")

else:
    print("invalid")
```

```python
subject1 = float(input("enter marks for subject 1: "))
subject2 = float(input("enter marks for subject 2: "))
subject3 = float(input("enter marks for subject 3: "))
subject4 = float(input("enter marks for subject 4: "))
subject5 = float(input("enter marks for subject 5: "))

print("grand total: 500")
total = subject1+subject2+subject3+subject4+subject5


average = total /5
print("total obtained by student: ",total)
print("average obtained by student: ",average)

if average >=80 and average <=100:
    # print("distinction")
    res = "distinction"

elif average >=60 and average<=79.9:
    # print("1st division")
    res = "1st division"

elif average >=50 and average<=59.9:
    # print("2nd division")
    res = "2nd division"

elif average >=40 and average<=49.9:
    # print("3rd division")
    res = "3rd division"

elif average<=39:
    print("fail")

else:
    print("invalid marks")

print("Your result is: ",res)
```
